````python
def bisection_method(f, a, b, tol=1e-6, max_iter=100):
    """
    Bisection method for root finding.
    """
    # TODO: Check if f(a)*f(b) < 0
    # TODO: Perform iterations
    # TODO: Return approximate root
    if f(a)*f(b) >= 0:
        print("Invalid interval!")
        return None

    while (b - a)/2 > tol:
        c = (a + b)/2

        if f(c) == 0:
            break
        elif f(a)*f(c) < 0:
            b = c
        else:
            a = c

    return (a + b)/2

````

````python
def false_position_method(f, a, b, tol=1e-6, max_iter=100):
    """
    False Position (Regula Falsi) method.
    """
    # TODO: Implement method
    if f(a)*f(b) >= 0:
        print("Invalid interval!")
        return None

    c = a
    while True:
        c_old = c
        c = (a*f(b) - b*f(a)) / (f(b) - f(a))

        if abs(c - c_old) < tol:
            break

        if f(a)*f(c) < 0:
            b = c
        else:
            a = c

    return c
````

````python
def newton_raphson_method(f, df, x0, tol=1e-6, max_iter=100):
    """
    Newton–Raphson method.
    """
    # TODO: Implement iteration formula
    x = x0

    for i in range(max_iter):
        x_new = x - f(x)/df(x)

        if abs(x_new - x) < tol:
            return x_new

        x = x_new

    print("Did not converge")
    return None
````

````python
def secant_method(f, x0, x1, tol=1e-6, max_iter=100):
    """
    Secant method.
    """
    # TODO: Implement secant update
    for i in range(max_iter):
        if f(x1) - f(x0) == 0:
            print("Division by zero error!")
            return None

        x2 = x1 - f(x1)*(x1 - x0)/(f(x1) - f(x0))

        if abs(x2 - x1) < tol:
            return x2

        x0, x1 = x1, x2

    print("Did not converge")
    return None

    
````
