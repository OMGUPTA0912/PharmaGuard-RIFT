# Problem 1: Gauss Elimination Method

Solve the following system of linear equations using the **Gauss Elimination Method**:


\begin{aligned}
4x - y + z &= -5 \\
2x + 2y + 3z &= 10 \\
5x - 2y + 6z &= 1
\end{aligned}
---

## Task

- Form the augmented matrix  
- Apply forward elimination  
- Use back substitution to find \( x, y, z \)

```python
import numpy as np

# TODO: Define coefficient matrix A
A = np.array([
    [4,-1,1],
    [2,2,3],
    [5,-2,6]
], dtype=float)

# TODO: Define constant vector b
b = np.array([-5,10,1], dtype=float)

# TODO: Number of equations
n = 3

# Forward Elimination
for i in range(n):
    # TODO: Pivot element
    pivot = A[i][i]
    
    for j in range(i+1, n):
        # TODO: Compute elimination factor
        factor = (A[j][i]/pivot)
        
        # TODO: Update rows of A
        A[j] = A[j]-factor*A[i]
        
        # TODO: Update vector b
        b[j] = b[j]- factor*b[i]

# Back Substitution
x = np.zeros(n)
#print(A)
#print(b)
for i in range(n-1, -1, -1):
    # TODO: Compute solution component
    x[i] = (b[i] - np.dot(A[i][i+1:n], x[i+1:n])) / A[i][i]

# TODO: Print solution
print("Solution (Gauss Elimination):", x)
```

# Problem 2: Gauss–Jordan Method

Solve the following system of equations using the **Gauss–Jordan Method**:


\begin{aligned}
2x + y + z &= 10 \\
3x + 2y + 3z &= 18 \\
x + 4y + 9z &= 16
\end{aligned}


## Task

- Construct the augmented matrix  
- Convert it to reduced row echelon form  
- Extract the solution directly  

```python
import numpy as np

# TODO: Define augmented matrix [A | b]
aug = np.array([
    [2,1,1,10],
    [3,2,3,18],
    [1,4,9,16]
], dtype=float)

n = len(aug)  # number of equations

# Gauss–Jordan Elimination
for i in range(n):
    # TODO: Normalize pivot row
    pivot = aug[i][i]
    aug[i] = aug[i]/pivot
    
    for j in range(n):
        if j != i:
            # TODO: Eliminate other entries
            factor = aug[j][i]
            aug[j] = aug[j]-aug[i]*factor

# TODO: Extract solution
solution = aug[:, -1]

print("Solution (Gauss–Jordan):", solution)
```
