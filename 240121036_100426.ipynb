# Problem 1: Euler’s Method

Consider the differential equation:

$$\frac{dy}{dx} = x + y, \quad y(0) = 1$$

### Tasks:
1. Implement Euler’s method to solve the differential equation.
2. Compute $y(0.2)$ using:
   - $h = 0.05$
   - $h = 0.1$



```python
def euler(h):
    x=0
    y=1
    while x<0.2:
        y=y+h*(x+y)
        x+=h
    return y

y1=euler(0.05)
y2=euler(0.1)
print(f"h={0.05}, y1={y1:.6f}")
print(f"h={0.1}, y1={y2:.6f}")
```
