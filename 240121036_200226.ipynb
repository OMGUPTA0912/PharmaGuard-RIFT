The third order polynomial is given by:

P3(x) = b0 + b1(x − x0) + b2(x − x0)(x − x1) + b3(x − x0)(x − x1)(x − x2)

## Given Data

| x | 1 | 4 | 6 | 5 |
|---|---|---|---|---|
| f(x) | 0 | 1.3863 | 1.7918 | 1.6094 |

---


## Estimation
Using the polynomial, estimate ln(2).

---

##  Programming Task

### Algorithm (Step-by-Step Procedure)

1. Input the given data points in two arrays: one for ( x ) values and one for corresponding ( y = f(x) ) values.

2. Create a divided difference table and store the function values in the first column.

3. Compute all order divided differences using the standard divided difference formula.

6. Construct the Newton interpolation polynomial using the calculated divided difference coefficients.

7. Substitute ( x = 2 ) into the polynomial to estimate ( \ln(2) ).

8. Print the estimated value.

9. Compare the estimated value with the actual value of ( \ln(2) ).




````python
# Write your Newton interpolation code here

import numpy as np
import math

x = np.array([1, 4, 6, 5], dtype=float)
y = np.array([0, 1.3863, 1.7918, 1.6094], dtype=float)

n = len(x)


dd = np.zeros((n, n))
dd[:, 0] = y


for j in range(1, n):
    for i in range(n - j):
        dd[i][j] = (dd[i+1][j-1] - dd[i][j-1]) / (x[i+j] - x[i])


def newton_interpolation(value):
    result = dd[0][0]
    product = 1.0
    
    for i in range(1, n):
        product *= (value - x[i-1])
        result += dd[0][i] * product
        
    return result


ln2_estimate = newton_interpolation(2)

print("Estimated ln(2) using 3rd order Newton polynomial:", ln2_estimate)
print("actual value of ln2 is", math.log(2) )
print("difference between actual and estimated is:", math.log(2)-ln2_estimate)
````

# Problem 2: Lagrange Interpolation

The Lagrange interpolating polynomial is given by:

Pn(x) = Σ f(xi) Li(x)

where

Li(x) = Π (x − xj) / (xi − xj),  for j ≠ i

## Given Data

| x | 1 | 4 | 6 |
|---|---|---|---|
| ln(x) | 0 | 1.3863 | 1.7918 |

---



##  Programming Task
### Algorithm (Step-by-Step Procedure) — Lagrange Interpolation

1. Input the given data points in two arrays: one for ( x ) values and one for corresponding ( y = f(x) ) values.

2. Choose the order of interpolation (first order or second order) by selecting the required number of data points.

3. For each data point, construct the corresponding Lagrange basis polynomial.

4. For each basis polynomial, compute the product term by multiplying the factors involving differences of ( x ) values.

5. Multiply each Lagrange basis polynomial by its corresponding function value.

6. Sum all the terms to construct the Lagrange interpolation polynomial.

7. Substitute ( x = 2 ) into the polynomial to estimate ( ln(2) ).

8. Print the estimated value.

9. Compare the estimated value with the actual value of ( ln(2) ).


###  Programming Task
Write a Python program for first order Lagrange interpolation.



````python
import numpy as np

x1 = np.array([1, 4], dtype=float)
y1 = np.array([0, 1.3863], dtype=float)

def lagrange(x_data, y_data, value):
    n = len(x_data)
    total = 0
    
    for i in range(n):
        term = y_data[i]
        for j in range(n):
            if j != i:
                term *= (value - x_data[j]) / (x_data[i] - x_data[j])
        total += term
        
    return total

ln2_first_order = lagrange(x1, y1, 2)
print("Estimated ln(2) using 1st order Lagrange:", ln2_first_order)
print("actual value of ln2 is", math.log(2) )
print("difference between actual and estimated is:", math.log(2)-ln2_first_order)

````
