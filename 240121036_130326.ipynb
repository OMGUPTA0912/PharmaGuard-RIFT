## Problem 1
Estimate the derivative of

f(x) = -0.1x^4 - 0.15x^3 - 0.5x^2 - 0.25x + 1.2

using step sizes:
h = 0.5, 0.25, 0.125, 0.01

Use:
- Forward difference
- Backward difference
- Central difference

```python
# Define the function f(x)
def f(x):
    return -(0.1)*pow(x,4)-(0.15)*pow(x,3)-(0.5)*pow(x,2)-(0.25)*x+1.2
```

```python
# Choose the point where derivative is evaluated
x = 0
```

```python
# Step sizes
h_values = [0.5, 0.25, 0.125, 0.01]
```

```python
# Forward difference
def forward_diff(f, x, h):
    return (f(x+h) - f(x)) / h
```

```python
# Backward difference
def backward_diff(f, x, h):
    return (f(x)- f(x-h)) / h
```

```python
# Central difference
def central_diff(f, x, h):
    return (f(x+h)- f(x-h)) / (2*h)
```

```python
# Compute derivatives for different h
for h in h_values:
    fwd = forward_diff(f, x, h)
    bwd = backward_diff(f, x, h)
    cen = central_diff(f, x, h)

    print("h =", h)
    print("Forward:", fwd)
    print("Backward:", bwd)
    print("Central:", cen)
    print()
```

## Problem 2
Given data:

| Time (s) | 5 | 6 | 7 | 8 | 9 |
|---------|---|---|---|---|---|
| Distance (m) | 10 | 15 | 22 | 35 | 50 |

Find velocity and acceleration at:
- t = 7 s
- t = 9 s

```python
# Store the data
time = [5, 6, 7, 8, 9]
distance = [10, 15, 22, 35, 50]
```

```python
# Step size
h = 1
```

```python
# Velocity at t = 7 using central difference
v7 = (distance[3] - distance[1]) / (2*h)
print("Velocity at t=7:", v7)
```

```python
# Velocity at t = 9 using backward difference
v9 = (distance[4] - distance[3]) / h
print("Velocity at t=9:", v9)
```

```python
# Acceleration at t = 7
a7 = ( distance[3]- 2*distance[2]+ distance[1]) / (h**2)
print("Acceleration at t=7:", a7)
```

```python
# Acceleration at t = 9 (use backward difference form)
a9 = ( distance[4]- 2*distance[3]+ distance[2]) / (h**2)
print("Acceleration at t=9:", a9)
```
