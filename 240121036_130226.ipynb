## Problem 1: Least Squares Fit

Fit the curve:

y = a x + b x^{-0.5}

Given data:

x: 0.2, 0.3, 0.5, 1.0, 2.0
y: 16, 14, 11, 6, 3

### Mathematical Procedure:
1. Compute the following quantities:
   - x^{-0.5}
   - x^2
   - x · x^{-0.5}
   - x^{-1}
   - yx
   - yx^{-0.5}

2. Compute the required summations.
3. Form the normal equations:

   Σ(yx) = a Σ(x²) + b Σ(x x^{-0.5})
   Σ(yx^{-0.5}) = a Σ(x x^{-0.5}) + b Σ(x^{-1})

4. Solve the two linear equations manually for a and b.

5. Plot the least squares fitted curve along with the given data points.


### Step 1: Import NumPy


```python
import numpy as np
```

### Step 2: Enter the given data


```python
x = [0.2, 0.3, 0.5, 1.0, 2.0]
y = [16, 14, 11, 6, 3]
```

### Step 3: Compute required quantities
Compute and print:
- x^(-0.5)
- x^2
- x * x^(-0.5)
- x^(-1)
- y*x
- y*x^(-0.5)


```python
n=len(x)
m=len(y)
for i in range(n):
    print("x^(-0.5)=",x[i]**(-0.5))
for i in range(n):
    print("x^2=",x[i]**(2))
for i in range(n):
    print("x*x^(-0.5)=",x[i]*(x[i]**(-0.5))) 
for i in range(n):
    print("x^(-1)=",x[i]**(-1))    
for i in range(m):
    for j in range(n):
    
       print("y^x=",y[i]*x[j])
for i in range(m):
    for j in range(n):
    
       print("y^x(-0.5)=",y[i]*(x[j]**(-0.5)))    
```

### Step 4: Compute required summations
Compute:
- Σ(x²)
- Σ(x x^{-0.5})
- Σ(x^{-1})
- Σ(yx)
- Σ(yx^{-0.5})


```python
n=len(x)


sum_x2 = sum(xi**2 for xi in x)
print(sum_x2)
sum_x3=sum(xi*xi**(-0.5) for xi in x)
print(sum_x3)
sum_x4=sum(xi**(-1) for xi in x)
print(sum_x4)

sum_xy = sum(x[i]*y[i] for i in range(n))
print(sum_xy)
sum_xy2 = sum(y[i]*(x[i]**(-0.5)) for i in range(n))
print(sum_xy2)      
```

### Step 5: Form the Normal Equations
Write the two equations clearly in the form:

A11 a + A12 b = B1
A21 a + A22 b = B2

Print the numerical values of A11, A12, A22, B1, B2.


```python
A=np.array([[sum_x2,sum_x3],[sum_x3,sum_x4]])
B=np.array([sum_xy,sum_xy2])
print(A)
print(B)
```

### Step 6: Solve the 2×2 system manually
Use algebra (substitution or determinant method):

a = (B1*A22 - B2*A12)/(A11*A22 - A12*A21)
b = (A11*B2 - A21*B1)/(A11*A22 - A12*A21)

Print values of a and b.


```python
a=(B[0]*A[1][1]-B[1]*A[0][1])/(A[0][0]*A[1][1]-A[0][1]*A[1][0])
print(a)
b = (A[0][0]*B[1] - A[1][0]*B[0])/(A[0][0]*A[1][1] - A[0][1]*A[1][0])
print(b);
```

### Step 6: Plot the least squares fitted curve

```python
import numpy as np
import matplotlib.pyplot as plt




x_fit = np.linspace(min(x), max(x), 200)
y_fit = a*x_fit + b*(x_fit**(-0.5))

# Plot
plt.scatter(x, y)
plt.plot(x_fit, y_fit)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Least Squares Fitted Curve")
plt.show()
```

## Problem 2: Linear Interpolation

Given:

x: 1, 4, 9, 16
f(x): 1, 2, 3, 4

To estimate √2:

1. Identify the interval containing x = 2.
2. Apply the linear interpolation formula:

   f(x) = f(x0) + ((x - x0)/(x1 - x0)) (f(x1) - f(x0))

3. Print the approximate value.


### Step 1: Store interpolation data


```python
x = [1, 4, 9, 16]
fx = [1, 2, 3, 4]
```

### Step 2: Identify correct interval for x = 2


```python
value = 2

for i in range(len(x) - 1):
    if x[i] <= value <= x[i + 1]:
        x0 = x[i]
        x1 = x[i + 1]
        f0 = fx[i]
        f1 = fx[i + 1]
        break
```

### Step 3: Apply linear interpolation formula and print result


```python



result = f0 + ((value - x0) / (x1 - x0)) * (f1 - f0)

print("Approximate value of sqrt(2):", result)
```
