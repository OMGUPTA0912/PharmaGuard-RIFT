## 🔹 Problem 2: Falling Parachutist

The velocity of a falling parachutist is given by:

$$
v(t) = \frac{gm}{c} \left(1 - e^{-(c/m)t} \right)
$$

where:

- $v(t)$ is the velocity (m s$^{-1}$)  
- $g = 9.8 \, \text{m s}^{-2}$ (gravitational acceleration)  
- $m = 68.1 \, \text{kg}$ (mass of the parachutist)  
- $c = 12.5 \, \text{kg s}^{-1}$ (drag coefficient)

---

After a time $t = 8 \, \text{s}$, the distance fallen $d$ (in cm) is:

$$
d = \int_{0}^{t} v(t)\, dt = \frac{gm}{c} \int_{0}^{t} \left(1 - e^{-(c/m)t} \right) dt
$$

---

### 🔸 Task

Using the **trapezoidal rule**, calculate $d$ for:

- $(n, h) = (20, 0.5)$  
- $(n, h) = (100, 0.1)$  
- $(n, h) = (500, 0.02)$  
- $(n, h) = (1000, 0.01)$  

---

### 🔸 Requirements

- Estimate the **percentage error** for each case  
- $n$ = number of segments  
- $h$ = step size  

---

```python
import math
def exact_f1(a, b):
    return (53.39)*(b+((math.exp(-1*(0.1838)*(b)))/(0.1838)))- (53.39)*(a+((math.exp(-1*(0.1838)*(a)))/(0.1838)))
def f(x):
    return (53.39)*(1-math.exp(-1*(0.1838)*(x)))
def trapezoidal(f, a, b, n):
    h = (b - a) / n
    result = 0.5 * (f(a) + f(b))

    for i in range(1, n):
        result += f(a + i * h)

    return result * h
a=0
b=10

trap1 = trapezoidal(f, a, b, 20)
trap2 = trapezoidal(f, a, b, 100)
trap3 = trapezoidal(f, a, b, 500)
trap4 = trapezoidal(f, a, b, 1000)
etrap1=exact_f1(0,10)
print(trap1)
print("%error1",abs(1-trap1/etrap1)*100)
print(trap2)
print("%error2",abs(1-trap2/etrap1)*100)
print(trap3)
print("%error3",abs(1-trap3/etrap1)*100)
print(trap4)
print("%error4",abs(1-trap4/etrap1)*100)

```
