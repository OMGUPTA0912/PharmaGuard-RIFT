
Evaluate the integral:

I = ∫₀^{π/2} (6 + 3 cos x) dx

Tasks:

```python
import numpy as np

# Define function
def f(x):
    return 6 + 3*np.cos(x)
```

## (a) Analytical solution

```python
import numpy as np

def ef(x):
    return 6*x + 3*np.sin(x)
I_exact=ef((np.pi)/2)-ef(0)
print(I_exact)
```

## (b) Trapezoidal Rule (n=2)

```python

import numpy as np
def trapezoidal(f, a, b, n):
    h = (b - a) / n
    result = 0.5 * (f(a) + f(b))

    for i in range(1, n):
        result += f(a + i * h)

    return result * h

a=0
b=(np.pi)/2
n = 2
trap = trapezoidal(f, a, b, n)
print(trap)
```

## (c) Simpson’s 1/3 Rule

```python
import numpy as np
def simpson(f,a, b, n):
    
    
    h = (b - a) / n
    s = f(a) + f(b)
    
    for i in range(1, n):
        if i % 2 == 0:
            s += 2 * f(a + i*h)
        else:
            s += 4 * f(a + i*h)
    
    return (h/3) * s

a=0
b=(np.pi)/2
n = 2
simp = simpson(f, a, b, n)
print(simp)
```
