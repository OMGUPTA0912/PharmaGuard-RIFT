## Problem 1: Numerical Integration using Trapezoidal Rule

Evaluate the integral:

$$
I = \int_a^b (x^3 + 1)\, dx
$$


### Tasks:

1. Compute the integral using the **Trapezoidal Rule** for the following intervals:
   - (a) \([1, 2]\)
   - (b) \([1, 1.5]\)

2. Estimate the **truncation error** in each case.

3. Compute the **exact analytical value** of the integral.

4. Compare the numerical results with the exact values.


````python
import math
def exact_f1(a, b):
    return (b**4)/4 + b - ((a**4)/4 + a)

def trapezoidal(f, a, b, n):
    h = (b - a) / n
    result = 0.5 * (f(a) + f(b))

    for i in range(1, n):
        result += f(a + i * h)

    return result * h

def f(a):
    return a**3+1

a = 1
b=2
n = 1
trap = trapezoidal(f, a, b, n)
exact = exact_f1(a, b)
error = abs(exact - trap)

print("Trapezoidal", trap) 
print("Exact", exact)
print("error", error)

a= 1
b=1.5
n = 1
trap2 = trapezoidal(f, a, b, n)
exact2 = exact_f1(a, b)
error2 = abs(exact - trap)
print("Trapezoidal", trap2) 
print("Exact", exact2)
print("error", error2)
````
